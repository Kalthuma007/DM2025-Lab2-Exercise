# **Competition Code**

## 1. Preprocessing Steps

```python
### Add the code related to the preprocessing steps in cells inside this section

# Load competition files
import pandas as pd
import re

split_info = pd.read_csv('data/dm-lab-2-private-competition/data_identification.csv')
emotion     = pd.read_csv('data/dm-lab-2-private-competition/emotion.csv')
sample_sub  = pd.read_csv('data/dm-lab-2-private-competition/samplesubmission.csv')
posts_raw   = pd.read_json('data/dm-lab-2-private-competition/final_posts.json')

# Flatten nested JSON
posts = pd.json_normalize(posts_raw['root'])

# Extract correct id/text columns
posts = posts[['_source.post.post_id', '_source.post.text']]
posts = posts.rename(columns={'_source.post.post_id': 'id', '_source.post.text': 'text'})

# Merge all data starting from split_info
full_data = split_info.merge(posts, on='id', how='left') \
                      .merge(emotion[['id', 'emotion']], on='id', how='left')

# Split into train and test
train = full_data[full_data['split'] == 'train'].copy()
test  = full_data[full_data['split'] == 'test'].copy()

# Clean text
def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", '', text)
    text = re.sub(r"@\w+|#", '', text)
    text = re.sub(r"[^\w\s]", ' ', text)
    text = re.sub(r"\s+", ' ', text).strip()
    return text

train.loc[:, 'clean_text'] = train['text'].apply(clean_text)
test.loc[:,  'clean_text'] = test['text'].apply(clean_text)

import matplotlib.pyplot as plt
import os

os.makedirs("pics", exist_ok=True)

# Plot distribution of emotions in training set
plt.figure(figsize=(8,6))
train['emotion'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Distribution of Emotions in Training Set")
plt.xlabel("Emotion")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("pics/emotion_distribution.png")  # save into pics folder
plt.show()
```

## 2. Feature Engineering Steps

```python
### Add the code related to the feature engineering steps in cells inside this section
import matplotlib.pyplot as plt
import numpy as np
import os

os.makedirs("pics", exist_ok=True)

# plot top 20 TF-IDF features by average weight
feature_names = vectorizer.get_feature_names_out()
avg_weights = X_train.mean(axis=0).A1
top_idx = np.argsort(avg_weights)[-20:]

plt.figure(figsize=(10,6))
plt.barh(range(20), avg_weights[top_idx], align='center', color='orange')
plt.yticks(range(20), [feature_names[i] for i in top_idx])
plt.xlabel("Average TF-IDF Weight")
plt.title("Top 20 TF-IDF Features")
plt.tight_layout()
plt.savefig("pics/tfidf_features.png")  # save into pics folder
plt.show()


# TF-IDF vectorization
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TF-IDF with unigrams and bigrams, limit to 2000 features
vectorizer = TfidfVectorizer(max_features=2000, ngram_range=(1,2))

# Fit on training data and transform both train and test sets
X_train = vectorizer.fit_transform(train['clean_text'])
X_test  = vectorizer.transform(test['clean_text'])

# Extract labels for training
y_train = train['emotion']

# Quick check: print shape of feature matrices
print("Training features shape:", X_train.shape)
print("Test features shape:", X_test.shape)
print("Number of labels:", len(y_train))
```

## 3. Model Implementation Steps

```python
### Add the code related to the model implementation steps in cells inside this section
# Train logistic regression model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Predict on test set
preds = model.predict(X_test)

# Build submission file
pred_df = test[['id']].copy()
pred_df['label'] = preds

submission = sample_sub.merge(pred_df, on='id', how='left')

# Fill any missing predictions with mode of y_train
if submission['label'].isna().any():
    default = y_train.mode().iloc[0]
    submission['label'] = submission['label'].fillna(default)

submission.to_csv('submission.csv', index=False)
print("Submission file created:", submission.shape)
print(submission.head())

```

# 4. Cross‑Validation (Optional)


```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import f1_score
import numpy as np

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores = []

for train_idx, val_idx in skf.split(X_train, y_train):
    X_tr, X_val = X_train[train_idx], X_train[val_idx]
    y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]
    
    clf = LogisticRegression(max_iter=1000, class_weight='balanced')
    clf.fit(X_tr, y_tr)
    y_pred = clf.predict(X_val)
    scores.append(f1_score(y_val, y_pred, average='macro'))

print("CV macro F1:", np.mean(scores), "±", np.std(scores))
```
